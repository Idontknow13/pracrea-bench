```python
import matplotlib.pyplot as plt
import benchmark

values = []
with open("test.data") as testdata:
    values = [line.strip() for line in testdata.readlines()]
```

```python
# Copy cell for every language test
lang = "rust-debug"
bubble_bench_debug = benchmark.Benchmark(values, benchmark.get_algorithms("quadratic"))
bubble_bench_debug.run_with(benchmark.get_binpath(lang))
```

```python
# Copy cell for every language test
lang = "rust-release"
bubble_bench_release = benchmark.Benchmark(values, benchmark.get_algorithms("quadratic"))
bubble_bench_release.run_with(benchmark.get_binpath(lang))
```

```python
x = bubble_bench_debug.valuesizes #! DO NOT REDECLARE
# Aggregate all the results here
bubble_rust_debug = bubble_bench_debug.results["bubble-sort"]
bubble_rust_release = bubble_bench_release.results["bubble-sort"]
```

```python
plt.title("The Runtime of Quadratic Time Algorithms")
plt.xlabel("Input Size")
plt.ylabel("Execution Time in ms")
# Add variables here
plt.plot(x, bubble_rust_debug, label="Bubble Sort (Rust: Debug)")
plt.plot(x, bubble_rust_release, label="Bubble Sort (Rust: Release)")
# Plotting
plt.legend()
plt.show()
```
